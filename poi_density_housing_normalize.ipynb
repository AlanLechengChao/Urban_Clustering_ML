```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
fishnet_poi_density = pd.read_excel('poi_fishnet.xlsx')
fishnet_poi_density
```

```python
import json
fishnet_town_map = {}
for i in range(2915):
    town_map = fishnet_poi_density[fishnet_poi_density.fid_fishne == i]
    f_town = town_map.drop_duplicates('town', 'first')
    f_town_list = np.array(f_town.town)
    fishnet_town_map[i] = f_town_list
fishnet_town_map


```

```python
fishnet_district_map = {}
for i in range(2915):
    district_map = fishnet_poi_density[fishnet_poi_density.fid_fishne == i]
    f_district = district_map.drop_duplicates('district_12', 'first')
    f_district_list = np.array(f_district.district_12)
    fishnet_district_map[i] = f_district_list
fishnet_district_map
```

```python
# fishnet_poi_density[fishnet_poi_density.town == '豫园街道']
towns = fishnet_poi_density.drop_duplicates('town', 'first')
towns = np.array(towns.town)
towns
```

```python
town_fishnet_map = {}
for t in towns:
    f = fishnet_poi_density[fishnet_poi_density.town == t]
    town_f = f.drop_duplicates('fid_fishne', 'first')
    town_f_list = np.array(town_f.fid_fishne, dtype=np.int16)
    print(t, town_f_list)
    town_fishnet_map[t] = town_f_list
```

```python
districts = fishnet_poi_density.drop_duplicates('district_12', 'first')
districts = np.array(districts.district_12)
districts
```

```python
district_fishnet_map = {}

for d in districts:
    # print(d)
    f = fishnet_poi_density[fishnet_poi_density.district_12 == d]
    # print(f)
    district_f = f.drop_duplicates('fid_fishne', 'first')
    district_f_list = np.array(district_f.fid_fishne, dtype=np.int16)
    print(d, district_f_list)
    district_fishnet_map[d] = district_f_list
```

```python
each_fishnet_density = pd.read_csv('fishnet_density.csv')

each_fishnet_poi_count = each_fishnet_density['不加权POI合计']
```

```python
each_fishnet_density
```

```python
town_result = pd.DataFrame(columns=['town', 'mean', 'std', 'f_list', 'f_len'])
for t, f_list in town_fishnet_map.items():
    f_density_list = []
    for f in f_list:
        f_density_list.append(each_fishnet_poi_count[f])
    mean = np.mean(f_density_list)
    std = np.std(f_density_list)
    town_result = town_result.append({'town': t, 'mean': mean, 'std': std, 'f_list':f_list, 'f_len': len(f_list)}, ignore_index=True)
town_result
```

```python
district_result = pd.DataFrame(columns=['district', 'mean', 'std', 'f_list', 'f_len'])
for t, f_list in district_fishnet_map.items():
    f_density_list = []
    for f in f_list:
        f_density_list.append(each_fishnet_poi_count[f])
    mean = np.mean(f_density_list)
    std = np.std(f_density_list)
    district_result = district_result.append(
        {'district': t, 'mean': mean, 'std': std, 'f_list': f_list, 'f_len': len(f_list)}, ignore_index=True)
district_result
```

```python
each_fishnet_density = each_fishnet_density.assign(normalized_density_town='', normalized_density_district='')
```

```python
import random
for l in town_result.f_list:
    for f in l:
        if each_fishnet_density.normalized_density_town[f] == '':
            town_mapped_from_fishnet = fishnet_town_map[f]
            town_mapped_from_fishnet = town_mapped_from_fishnet[random.randint(0, len(town_mapped_from_fishnet)-1)]
            try:
                mean = town_result[town_result.town == town_mapped_from_fishnet]['mean'].values[0]
                std = town_result[town_result.town ==
                              town_mapped_from_fishnet]['std'].values[0]
            except:
                continue
            # print(mean, std)
            each_fishnet_density.normalized_density_town[f] = float((each_fishnet_poi_count[f] - mean) / std)
for l in district_result.f_list:
    for f in l:
        if each_fishnet_density.normalized_density_district[f] == '':
            district_mapped_from_fishnet = fishnet_district_map[f]
            
            district_mapped_from_fishnet = district_mapped_from_fishnet[random.randint(
                0, len(district_mapped_from_fishnet)-1)]
            # print(district_mapped_from_fishnet)
            try:
                mean = district_result[district_result.district ==
                                   district_mapped_from_fishnet]['mean'].values[0]
                std = district_result[district_result.district ==
                                  district_mapped_from_fishnet]['std'].values[0]
            except:
                continue
            each_fishnet_density.normalized_density_district[f] = (each_fishnet_poi_count[f] - mean) / std
each_fishnet_density
```

```python
housing = pd.read_excel('housing_f_t.xlsx')
housing
```

```python
housing_result_town = pd.DataFrame(
    columns=['town', 'mean', 'std', 'f_list', 'f_len'])
for t, f_list in town_fishnet_map.items():
    f_housing_list = []
    for f in f_list:
        f_housing_list += housing[housing.fid_fishne == f]['单价_元'].to_list()
    # print(f_housing_list)
    mean = np.mean(f_housing_list)
    std = np.std(f_housing_list)
    housing_result_town = housing_result_town.append(
        {'town': t, 'mean': mean, 'std': std, 'f_list': f_list, 'f_len': len(f_list)}, ignore_index=True)
housing_result_town
```

```python
housing_result_district = pd.DataFrame(
    columns=['district', 'mean', 'std', 'f_list', 'f_len'])
for d, f_list in district_fishnet_map.items():
    f_housing_list = []
    for f in f_list:
        f_housing_list += housing[housing.fid_fishne == f]['单价_元'].to_list()
    # print(f_housing_list)
    mean = np.mean(f_housing_list)
    std = np.std(f_housing_list)
    housing_result_district = housing_result_district.append(
        {'district': d, 'mean': mean, 'std': std, 'f_list': f_list, 'f_len': len(f_list)}, ignore_index=True)
housing_result_district
```

```python
each_fishnet_density = each_fishnet_density.assign(housing_town='', housing_district='')
each_fishnet_density
```

```python
housing_kriging = pd.read_excel('housing_kriging.xlsx')
housing_kriging
```

```python
count = 0
for i in range(2915):
    town_mapped_from_fishnet = fishnet_town_map[i]
    if len(town_mapped_from_fishnet) == 0:
        count += 1
        continue
    town_mapped_from_fishnet = town_mapped_from_fishnet[random.randint(
        0, len(town_mapped_from_fishnet)-1)]
    try:
        mean = housing_result_town[housing_result_town.town == town_mapped_from_fishnet]['mean'].values[0]
        std = housing_result_town[housing_result_town.town == town_mapped_from_fishnet]['std'].values[0]
    except:
        # count += 1
        continue
    # print(mean, std)
    try:
        each_fishnet_density.housing_town[i] = (each_fishnet_density['房价'][i] - mean) / std
    except:
        each_fishnet_density.housing_town[i] = (housing_kriging['fid_fishne'][i] - mean) / std
for i in range(2915):
    district_mapped_from_fishnet = fishnet_district_map[i]
    if len(district_mapped_from_fishnet) == 0:
        continue
    district_mapped_from_fishnet = district_mapped_from_fishnet[random.randint(
        0, len(district_mapped_from_fishnet)-1)]
    try:
        mean = housing_result_district[housing_result_district.district ==
                                   district_mapped_from_fishnet]['mean'].values[0]
        std = housing_result_district[housing_result_district.district ==
                                  district_mapped_from_fishnet]['std'].values[0]
    except:
        continue
    # print(mean, std)
    try:
        each_fishnet_density.housing_district[i] = (
            each_fishnet_density['房价'][i] - mean) / std
    except:
        each_fishnet_density.housing_town[i] = (
            housing_kriging['fid_fishne'][i] - mean) / std
# each_fishnet_density
count
```

```python
each_fishnet_density = each_fishnet_density[each_fishnet_density.normalized_density_town != '']
# each_fishnet_density = each_fishnet_density.replace('', -1)
each_fishnet_density.to_csv('norm_f_town_district.csv')
```

```python
each_fishnet_density_geo = gpd.GeoDataFrame(each_fishnet_density, geometry=gpd.points_from_xy(
    each_fishnet_density['经度'], each_fishnet_density['纬度']), crs='epsg:4326')
each_fishnet_density_geo
```

```python
each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='不加权POI合计')
plt.title('POI density, without normalize')

each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='加权合计POI')
plt.title('POI weighted density, without normalize')

each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='normalized_density_town')
plt.title('POI density normalized, town scale')
each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='normalized_density_district')
plt.title('POI density normalized, district scale')
each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='housing_town')
plt.title('housing normalized, town scale')
each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='housing_district')
plt.title('housing normalized, district scale')
each_fishnet_density_geo.plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='房价')
plt.title('housing, without normalize')
each_fishnet_density_geo[each_fishnet_density_geo['商铺人均消费'] != 0].plot(figsize=(10, 10), cmap=plt.get_cmap(
    'jet'), column='商铺人均消费')
plt.title('Dianping Average')
```
