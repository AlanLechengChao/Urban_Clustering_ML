```python
# ! pip install geopandas
from sklearn.cluster import KMeans
from sklearn.cluster import AffinityPropagation 
from sklearn.mixture import GaussianMixture
from sklearn import preprocessing
from sklearn.cluster import DBSCAN
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import io

# from google.colab import files
# uploaded = files.upload()

df = pd.read_csv("POI_Fishnet_for_training.csv",usecols=[1,2,4])
columns = list(df.columns)
df = df.fillna(0)
values = df.values
min_max_scaler = preprocessing.MinMaxScaler()
values_scaled = min_max_scaler.fit_transform(values)
df = pd.DataFrame(values_scaled, columns=columns)
df
```

```python
kmeans = KMeans(n_clusters=7, random_state=0, n_init=6)
kmeans.fit(df)

GM=GaussianMixture(n_components=7)
GM.fit(df)

# AP = AffinityPropagation()
# AP.fit(df)

db=DBSCAN(eps=0.3,min_samples=8).fit(df)
# a=set(db.labels_)
# print(a)
```

```python
#print(kmeans.labels_)

df['GaussianMixture_result'] = GM.predict(df)
df['Kmeans_result'] = kmeans.labels_
# df['AffinityPropagation_result'] = AP.labels_
df['db_result']=db.labels_

```

```python

plt.figure()
df.plot.scatter(x='POI_normalize', y='housing_price',
                     c='Kmeans_result', cmap=plt.get_cmap('jet'), colorbar=False, title='k means, 7 clusters')

```
